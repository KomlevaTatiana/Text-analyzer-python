```python
#Text Analyzer Project
#Beginner Python project for text analysis and visualization
import re
from collections import Counter
import matplotlib.pyplot as plt
def input_option():
  while True:
    choice = input('Choose input: A) type/paste  B) load .txt ').strip().upper()
    if choice in ('A', 'B'):
      return choice
    print('Please type only "A" or "B".')
def read_text(stop_phrase='the end'):
  lines=[]
  print("\nType/paste your text line by line.")
  print(f"Finish input by typing: {stop_phrase}")
  print("Empty line = new paragraph.\n")
  while True:
    line = input()
    if line.strip().upper() == stop_phrase.upper():
      break
    lines.append(line)
  return lines
def read_file(filepath):
  with open(filepath, 'r', encoding='utf-8') as f:
    text=f.read()
  return text.splitlines()
def lines_to_full_text(lines):
  return '\n'.join(lines)
def tokenize_words(text):
    t = text.lower()
    t = re.sub(r"(?<=\w)[-'](?=\w)", "", t)
    t = re.sub(r"[^\w\s]", " ", t)
    t = t.replace("_", " ")
    return t.split()
def count_paragraphs(lines):
  paragraphs = 0
  in_paragraph = False
  for line in lines:
    if line.strip()=='':
      in_paragraph = False
    else:
      if not in_paragraph:
        paragraphs+=1
        in_paragraph = True
  return paragraphs
def count_sentences(text):
  parts = re.split(r"[.!?]+", text)
  return sum(1 for p in parts if p.strip())
def build_stats(lines):
  text=lines_to_full_text(lines)
  words=tokenize_words(text)
  freq=dict(Counter(words))
  stats={
      'total_characters_with_spaces': len(text),
      'total_characters_without_spaces': len(''.join(text.split())),
      'total_words': len(words),
      'total_sentences': count_sentences(text),
      'total_paragraphs': count_paragraphs(lines),
      'total_unique_words': len(freq),
  }
  return stats, freq
def top_n_words(freq, n=10):
  return Counter(freq).most_common(n)
def print_results(stats, top10):
  print(f"Total characters (with spaces): {stats['total_characters_with_spaces']}")
  print(f"Total characters (without spaces): {stats['total_characters_without_spaces']}")
  print(f"Total words: {stats['total_words']}")
  print(f"Total sentences: {stats['total_sentences']}")
  print(f"Total paragraphs: {stats['total_paragraphs']}\n")

  print('Top 10 most common words:')
  for w, c in top10:
    print(f'{w}: {c}')
def plot_summary(stats):
  labels = ['Words', 'Sentences', 'Paragraphs', 'Unique words']
  values = [
      stats['total_words'],
      stats['total_sentences'],
      stats['total_paragraphs'],
      stats['total_unique_words']
  ]
  plt.bar(labels, values)
  plt.title('Text Summary')
  plt.ylabel('Count')
  plt.show()
def final():
  option = input_option()
  if option == 'A':
    lines = read_text('the end')
  else:
    filepath = input('Enter path to .txt file (e.g. /content/sample.txt): ').strip()
    try:
      lines= read_file(filepath)
    except FileNotFoundError:
      print('File not found. Check the path and try again.')
      return
    except UnicodeDecodeError:
      print('Cannot read file (encoding issue). Try saving it as UTF-8.')
      return
  stats, freq = build_stats(lines)
  top10 = top_n_words(freq, 10)
  print_results(stats, top10)
  plot_summary(stats)
final()




```

```python
#Version 2 with method class
import re
from collections import Counter
import matplotlib.pyplot as plt


class TextAnalyzerV2:
    def __init__(self, lines):
        self.lines = lines
        self.text = "\n".join(lines)
        self.stats = {}
        self.freq = {}

    def tokenize_words(self):
        t = self.text.lower()
        t = re.sub(r"(?<=\w)[-'](?=\w)", "", t)
        t = re.sub(r"[^\w\s]", " ", t)
        t = t.replace("_", " ")
        return t.split()

    def count_characters_with_spaces(self):
        return len(self.text)

    def count_characters_without_spaces(self):
        return len("".join(self.text.split()))

    def count_sentences(self):
        parts = re.split(r"[.!?]+", self.text)
        return sum(1 for p in parts if p.strip())

    def count_paragraphs(self):
        paragraphs = 0
        in_paragraph = False
        for line in self.lines:
            if line.strip() == "":
                in_paragraph = False
            else:
                if not in_paragraph:
                    paragraphs += 1
                    in_paragraph = True
        return paragraphs

    def analyze(self):
        words = self.tokenize_words()
        self.freq = dict(Counter(words))
        self.stats = {
            "total_characters_with_spaces": self.count_characters_with_spaces(),
            "total_characters_without_spaces": self.count_characters_without_spaces(),
            "total_words": len(words),
            "total_sentences": self.count_sentences(),
            "total_paragraphs": self.count_paragraphs(),
            "total_unique_words": len(self.freq),
        }
        return self.stats

    def top_n_words(self, n=10):
        return Counter(self.freq).most_common(n)

    def print_results(self, n_top=10):
        if not self.stats or not self.freq:
            self.analyze()

        print(f"Total characters (with spaces): {self.stats['total_characters_with_spaces']}")
        print(f"Total characters (without spaces): {self.stats['total_characters_without_spaces']}")
        print(f"Total words: {self.stats['total_words']}")
        print(f"Total sentences: {self.stats['total_sentences']}")
        print(f"Total paragraphs: {self.stats['total_paragraphs']}\n")

        print("Top 10 Most Common Words:")
        for w, c in self.top_n_words(n_top):
            print(f"{w}: {c}")

    def plot_summary(self):
        if not self.stats:
            self.analyze()

        labels = ["Words", "Sentences", "Paragraphs", "Unique words"]
        values = [
            self.stats["total_words"],
            self.stats["total_sentences"],
            self.stats["total_paragraphs"],
            self.stats["total_unique_words"],
        ]
        plt.bar(labels, values)
        plt.title("Text Summary")
        plt.ylabel("Count")
        plt.show()


def get_input_option_v2():
    while True:
        choice = input("Choose input: A) type/paste  B) load .txt  ").strip().upper()
        if choice in ("A", "B"):
            return choice
        print("Please type only A or B.")


def read_text_from_user_v2(stop_phrase="THE END"):
    lines = []
    print("\nType/paste your text line by line.")
    print(f"Finish input by typing: {stop_phrase}")
    print("Empty line = new paragraph.\n")

    while True:
        line = input()
        if line.strip().upper() == stop_phrase.strip().upper():
            break
        lines.append(line)
    return lines


def read_text_from_file_v2(filepath):
    with open(filepath, "r", encoding="utf-8") as f:
        text = f.read()
    return text.splitlines()


def main_v2():
    option = get_input_option_v2()

    if option == "A":
        lines = read_text_from_user_v2("THE END")
    else:
        filepath = input("Enter path to .txt file (e.g. /content/sample.txt): ").strip()
        try:
            lines = read_text_from_file_v2(filepath)
        except FileNotFoundError:
            print("File not found. Check the path and try again.")
            return
        except UnicodeDecodeError:
            print("Cannot read file (encoding issue). Try saving it as UTF-8.")
            return

    analyzer = TextAnalyzerV2(lines)
    analyzer.print_results(10)
    analyzer.plot_summary()


main_v2()

```
